```python
import cv2 as cv
import tkinter
import customtkinter
```

```python
def button_event():
    
    cap=cv.VideoCapture(0)

    med=cv.imread("C:/KSR/plus.jpg")
    fire=cv.imread("C:/KSR/fire.jpg")
    ex=cv.imread("C:/KSR/exit.jpg")
    hyd=cv.imread("C:/KSR/hydrant.jpg")
    phone=cv.imread("C:/KSR/phone.jpg")

    med=cv.resize(med,(64,64))
    fire=cv.resize(fire,(64,64))
    ex=cv.resize(ex,(64,64))
    hyd=cv.resize(hyd,(64,64))
    phone=cv.resize(phone,(64,64))

    med=cv.inRange(med,(89,91,149), (255,255,255))
    fire=cv.inRange(fire,(89,91,149),(255,255,255))
    ex=cv.inRange(ex,(89,91,149), (255,255,255))
    hyd=cv.inRange(hyd,(89,91,149), (255,255,255))
    phone=cv.inRange(phone, (89,91,149), (255,255,255))

    pr_s = ""
    s = ""
    while (True):
        ret,frame=cap.read()

        frameCopy=frame.copy()

        hsv=cv.cvtColor(frame,cv.COLOR_BGR2HSV)
        has=cv.blur(hsv,(5,5))
        mask=cv.inRange(hsv,(89,124,73),(255,255,255))
 

        mask=cv.erode(mask,None,iterations=2)
        mask=cv.dilate(mask, None, iterations=4)


        contours=cv.findContours(mask,cv.RETR_TREE, cv.CHAIN_APPROX_NONE)
        contours=contours[0]
        if contours:
            contours=sorted(contours,key=cv.contourArea,reverse=True)
            cv.drawContours(frame,contours,0,(255,0,255),3)


            (x,y,w,h)=cv.boundingRect(contours[0])
            frame = cv.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
            frame = cv.putText(frame, s, (x, y - 5),
                        cv.FONT_HERSHEY_SIMPLEX, 1, (36,255,12), 2) 
            cv.imshow("Rect",frame)

            roImg=frameCopy[y:y+h,x:x+w]
            cv.imshow("Detect", roImg)
            roImg=cv.resize(roImg,(64,64,))
            roImg=cv.inRange(roImg,(110,250,250),(255,255,255))

           
            roImg1=frameCopy[y:y+h,x:x+w]
            roImg1=cv.resize(roImg1,(64,64,))
            roImg1=cv.inRange(roImg1,(110,110,110),(255,255,255)) 
           

            roImg2=frameCopy[y:y+h,x:x+w]
            roImg2=cv.resize(roImg2,(64,64,))
            roImg2=cv.inRange(roImg2,(110,80,0),(255,255,255)) 





            med_val0, med_val1, med_val2 = 0, 0, 0
            fire_val0, fire_val1, fire_val2 = 0, 0, 0
            ex_val0, ex_val1, ex_val2 = 0, 0, 0
            hyd_val0, hyd_val1, hyd_val2 = 0, 0, 0 
            phone_val0, phone_val1, phone_val2 = 0, 0, 0

            for i in range(64):
                for j in range(64):
                    if roImg[i][j] == med[i][j]:
                        med_val0+=1
                    if roImg1[i][j] == med[i][j]:
                        med_val1+=1
                    if roImg2[i][j] == med[i][j]:
                        med_val2+=1

                    if roImg[i][j]==fire[i][j]:
                        fire_val0+=1
                    if roImg1[i][j]==fire[i][j]:
                        fire_val1+=1
                    if roImg2[i][j]==fire[i][j]:
                        fire_val2+=1

                    if roImg[i][j]==ex[i][j]:
                        ex_val0+=1
                    if roImg1[i][j]==ex[i][j]:
                        ex_val1+=1
                    if roImg2[i][j]==ex[i][j]:
                        ex_val2+=1


                    if roImg[i][j]==phone[i][j]:
                        phone_val0+=1
                    if roImg1[i][j]==phone[i][j]:
                        phone_val1+=1
                    if roImg2[i][j]==phone[i][j]:
                        phone_val2+=1

        if cv.waitKey(1)==ord("q"):
            break

        s = ''  
        if med_val0 > 3100 or med_val1 > 3100 or med_val2 > 3100:
            s = "First aid kit"
        if fire_val0 > 3600 or fire_val1 > 3600 or fire_val2 > 3600:
            s = "Fire extinguisher"
        if ex_val0 > 3050 or ex_val1 > 3050 or ex_val2 > 3050:
            s = "Exit"
        if phone_val0 > 3200 or phone_val1 > 3200 or phone_val2 > 3200:
            s = "Phone for use in case of fire"


        if s != pr_s:
            print(s)

        pr_s = s
        

    cap.release()
    cv.destroyAllWindows()
```

```python
customtkinter.set_appearance_mode("dark")
customtkinter.set_default_color_theme("dark-blue")

window = customtkinter.CTk()
window.title("Sign Recognition")
window.geometry('400x250')


label = customtkinter.CTkLabel(master=window, text="Для подключения камеры нажмите СТАРТ")
label.place(relx=0.5, rely=0.4, anchor=tkinter.CENTER)

button = customtkinter.CTkButton(master=window, text="СТАРТ", command=button_event)
button.place(relx=0.5, rely=0.6, anchor=tkinter.CENTER)

window.focus()
window.mainloop()
```
